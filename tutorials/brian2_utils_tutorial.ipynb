```python
#%%imports
from brian2 import ms
from plotly.subplots import make_subplots

from snnib import brian2_utils
from snnib.simulations import brian2_simulation
```

## Brian2

### Run some simulation

```python
#run a simulation
# brian2_simulation.check_stp()
net, t_sim, dt, params, parts = brian2_simulation.experiment(3)
ng_in = parts[0]    #input neuron group

```

### Get a neurons `SpikeMonitor`

```python
#obtain spike monitor of input group
sm_in = brian2_utils.get_spike_monitor(net, ng_in)
print(sm_in)
```

### Get a plot-able spiketrain 

```python

t, i = brian2_utils.get_spiketrains(sm_in, t_max=t_sim, n2plot=10)

fig = make_subplots(1,1, x_title="Time [ms]", y_title="Neuron ID")
fig.add_trace(dict(
    x=t / ms,
    y=i,
    marker=dict(
        color=i,
    ),
    mode="markers",

))
fig.show()
```
